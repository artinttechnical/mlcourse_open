**3. Какова доля граждан Германии (признак *native-country*)?**

```python
# Ваш код здесь
germans = data[data["native-country"] == "Germany"]["age"].count()
print(germans)
all_citizens = data["age"].count()
print(all_citizens)
print(germans / all_citizens)
```

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
mean_sm = data[data["salary"] == "<=50K"]["age"].mean()
std_sm = data[data["salary"] == "<=50K"]["age"].std()
mean_gr = data[data["salary"] == ">50K"]["age"].mean()
std_gr = data[data["salary"] == ">50K"]["age"].std()

print(mean_sm, std_sm, mean_gr, std_gr) 
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# Ваш код здесь
#pd.Series.astype?
mapping = {'Never-married' : "Single", 'Married-civ-spouse' : "Married", 'Divorced' : "Single",
       'Married-spouse-absent' : "Married", 'Separated' : "Single", 'Married-AF-spouse' : "Married",
       'Widowed' : "Single"}
#mar_data = data.set_index("marital-status")
#print(mapping.values())
#print(data["marital-status"].unique())
#mar_data.groupby(["sex", mapping, "salary"]).count()
#data[data["marital-status"].astype("str").startswith("Married")]
mar_data = data[(data["sex"] == "Male") & (data["salary"] == ">50K")]
mar_data.set_index("marital-status", inplace=True)
mar_data.groupby(mapping).count()
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
# Ваш код здесь
workers = data[(data["hours-per-week"] == 99) &(data["salary"] == "<=50K")]["age"].count()
topmanagers = data[(data["hours-per-week"] == 99) &(data["salary"] == ">50K")]["age"].count()
print(topmanagers / (topmanagers + workers))
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**
